```python
import specdal
import pandas as pd
import plotly.express as px
```

```python
sphere_df = specdal.Collection(name='int_sphere', measure_type='tgt_count',  directory='data/IntSphere')
sphere_df.data
```

```python
groups = sphere_df.groupby(separator='_', indices=[0, 1, 2])
group_names = list(groups.keys())
group_names
```

```python
# Species to process
target = "IS_EucRubida"
leaf = [i for i in sphere_df.data.columns if target in i]
leaf
```

```python
sphere_df.data[leaf]
```

```python
px.line(sphere_df.data[leaf], title=f'Reflectance of {target} (dn)', labels={'index': 'Wavelength (nm)', 'value': 'Reflectance (dn)'})
```

```python
# White reference and stray light meadsurements. Should be the closest in time to the leaf measurements. 

# White ref
wr_ref = [i for i in sphere_df.data.columns if "IS_WR_Ref" in i]
wr_trans = [i for i in sphere_df.data.columns if "IS_WR_Trans" in i]

# Stray light
sl_ref = [i for i in sphere_df.data.columns if "IS_SL_Ref" in i]
sl_trans = [i for i in sphere_df.data.columns if "IS_SL_Trans" in i]
```

```python
# White Reference
Ref_Rr = sphere_df.data[wr_ref].iloc[:,0]

# Stray Light
Ref_Id = sphere_df.data[sl_ref].iloc[:,0]

# Reference with leaf
Ref_Ir = sphere_df.data[leaf].iloc[:,0]

# Sample Reflectance
Ref_Is = sphere_df.data[leaf].iloc[:,3]

Rs = ((Ref_Is - Ref_Id)) / (Ref_Ir - Ref_Id) 

px.line(Rs, range_y=[0, 1], title=f'Reflectance of {target}', labels={'index':'Wavelength (nm)', 'value':'Reflectance'})
```

```python
# White Reference
Trans_Rr = sphere_df.data[wr_trans].iloc[:,0]

# Stray Light
Trans_Id = sphere_df.data[sl_trans].iloc[:,0]

# Reference with leaf
Trans_Ir = sphere_df.data[leaf].iloc[:,5]

# Sample Reflectance
Trans_Is = sphere_df.data[leaf].iloc[:,7]

T = ((Trans_Is - Trans_Id)) / (Trans_Ir - Trans_Id) 

px.line(Rs, range_y=[0, 1], title=f'Transmitance of {target}', labels={'index':'Wavelength (nm)', 'value':'Transmitance'})
```

```python
abs = 1 - T - Rs
px.line(Rs, range_y=[0, 1], title=f'Absorbance of {target}', labels={'index':'Wavelength (nm)', 'value':'Absorbance'})
```

```python
# plot all together
df = pd.DataFrame({'Reflectance': Rs, 'Transmitance': T, 'Absorbance': abs})
px.line(df, range_y=[0, 1], title=f'Reflectance, Transmitance and Absorbance of {target}', labels={'index':'Wavelength (nm)', 'value':'Reflectance'})
```

```python
df['Wavelength'] = df.index
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv("data/Output/" + str(target) + "_sphere.csv")
```
