```python
import specdal
import plotly.express as px
import pandas as pd
```

```python
probe_df = specdal.Collection(name='probe', measure_type='tgt_count',  directory='data/Probe')
```

```python
groups = probe_df.groupby(separator='_', indices=[1])
group_names = list(groups.keys())
```

```python
group_names
```

```python
target = 'EucObliqua'
```

```python
group = [col for col in probe_df.data if target in col]

probe_df.data[group]
```

```python
px.line(probe_df.data[group], title=f'Reflectance of {target} (dn)', labels={'index': 'Wavelength (nm)', 'value': 'Reflectance (dn)'})
```

```python
# Calculate the mean of the white reference from the first two columns
wr_mean = probe_df.data[group].iloc[:, :2].mean(axis=1)

# Calculate the mean of the rest of the columns, which is reflectance 
ref_mean = probe_df.data[group].iloc[:, 2:].mean(axis=1)

# Calculate absolute reflectance
abs_ref = ref_mean / wr_mean

# Plotting the absolute reflectance
px.line(abs_ref, title=f'Reflectance of {target}', labels={'index': 'Wavelength (nm)', 'value': 'Reflectance'})
```

```python
# Create DataFrame from ABS_REF
d = {'Reflectance': abs_ref}
df = pd.DataFrame(data=d)

# Reset index and set it as a regular column
df['Wavelength'] = df.index
df.reset_index(drop=True, inplace=True)
df
```

```python
# Save the DataFrame to a CSV file
df.to_csv("data/Output/" + str(target) + "_probe.csv")
```
